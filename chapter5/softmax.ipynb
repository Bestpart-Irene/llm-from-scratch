 $$S_i = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}$$

```python
#implementation of softmax

def softmax_torch(x, dim=-1):
    '''
    input:tenser
    output:probability
    '''

    # Subtract the maximum value for numerical stability
    max_val = torch.max(x, dim=dim, keepdim=True).values
    e_x = torch.exp(x-max_val)
    # Normalize by the sum of exponential
    output = e_x/torch.sum(e_x, dim=dim, keepdim=True)
    return output
```

```python
seq_len=10
x=torch.randn(seq_len)

print(f'x is {x}')
result = softmax_torch(x, dim=-1)
print(f'result: {result}')
print(f'sum of result is {torch.sum(result)}')
```

```python
batch_size, seq_len=16,10
x=torch.randn(batch_size, seq_len)

print(f'x is {x}')
result = softmax_torch(x, dim=-1)
print(f'result: {result}')
print(f'sum of result is {torch.sum(result,-1,keepdim=True)}')
```

```python
def softmax_temparature(x, temparature,dim=-1):

    max_val = torch.max(x, dim=dim, keepdim=True).values
    e_x = torch.exp((x-max_val)/temparature)
    output = e_x/torch.sum(e_x, dim=dim, keepdim=True)
    return output
```

```python
batch_size, seq_len=16,10
x=torch.randn(batch_size, seq_len)

print(f'x is {x}')
result = softmax_temparature(x, temparature=1.0, dim=-1)
print(f'result: {result}')
```
